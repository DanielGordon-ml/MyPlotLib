For the perpose of this example we will create sample data set. We create $4$ random columns 'A-D' and we create 'E' as linear combination of 'A' together with wihte noise using **np.random.randn**

```python
# dates 
dates = pd.date_range('2020-01-01', periods=100, freq='D')
dates.name = 'date'

# data
data = np.random.randn(100, 4)

# dataframe
df = pd.DataFrame(data, index=dates, columns=list('ABCD'))
df['E'] = 2*df['A']+1 + np.random.randn(df.shape[0])
```

Now, we can create the class object **TS_visual** and apply theme 'light'

```python
vis = TS_visual(df,'date')
vis.apply_theme('light')
```

Lets take a first look at the time series 'A'

```python
display(df['A'].head(2))

vis.line_plot('A')
```

We can now take a look at 'A' vs 'E' in both correlation and as lines:


```python
vis.correlation('A','E')
vis.lines_plot(['A','E'])
```

Using the full statistics plot we can now see the auto-correlation, partial auto-correlation, and time decomposition in one plot as following

```python
vis.full_statistics_plot('A',save=True)
```
